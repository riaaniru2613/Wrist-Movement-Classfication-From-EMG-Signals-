```python
import sys
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
```

```python
# Establishing the current work directory (cwd)
dir_path = '/content/drive/MyDrive/Major Project/Features Dataset'

# File_Names = []
# for (dir_path, dir_names, file_names) in os.walk(dir_path):
#     File_Names.extend(file_names)    
# print("Total files read is ", len(File_Names))

Files = []
for r, d, f in os.walk(dir_path):
    for file in f:
        if ".csv" in file:
            Files.append(os.path.join(r, file))
# filenames1 = Files[1:]
filenames = Files
```

```python
frames = []
frames1=[]
for file in filenames:  

  # print(file)
  file_data = pd.read_csv(file)
  # display(file_data)
  frames.append(file_data)
  
Data = pd.concat(frames, ignore_index = True)
print("Size of all the feature matrix combined is ", Data.shape)
```

```python
labels = np.array(Data['Output'])
# print(Data['Output'].value_counts())

Data= Data.drop('Output', axis = 1)

feature_list = list(Data.columns)

Data = np.array(Data)
```

```python
final_data=[]
temp=[]
final_label=[]
for i in range(len(Data)):
  
  for x in Data[i]:                     # get the current row 
    temp.append(x)
  
  if((i+1)%40 == 0):                    # single row in final_data corresponds to the entire movement(i.e 40 rows of Data)
    final_data.append(temp)
    final_label.append(labels[i])
    temp=[]
final_data=np.array(final_data)
final_labels=np.array(final_label)
```

```python
RF = RandomForestClassifier(n_estimators = 5000)
RF.fit(train_features, train_labels);
```

```python
predictions = RF.predict(test_features)

errors = abs(predictions - test_labels)

print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
mape = 100 * (errors / test_labels)

accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
